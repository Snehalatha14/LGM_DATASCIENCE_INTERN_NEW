# IMAGE TO PENCIL SKETCH

## IMPORT THE MODULE

```python
import cv2
```

## READING THE IMAGE AND DISPLAY

```python
#reading image and display
God_Image=cv2.imread("Lord.jpg")
cv2.imshow("Krishna", God_Image)
cv2.waitKey(0)
```

## CONVERTING THE IMAGE TO GREYSCALE IMAGE

```python
#converting original to greyscale
new_gImage=cv2.cvtColor(God_Image,cv2.COLOR_BGR2GRAY)
cv2.imshow("new_Lord", new_gImage)
cv2.waitKey(0)
```

## CONVERTING THE GREYSCALE IMAGE TO INVERTED IMAGE

```python
# invert the new greyscake image
new_inverted=255-new_gImage
cv2.imshow("Inverted image",new_inverted)
cv2.waitKey()
```

##  BLUR THE IMAGE

```python
# now blur the image using Gaussian function in Opencv
blurred=cv2.GaussianBlur(new_inverted,(21,21), 0)
cv2.imshow("blur",blurred)
cv2.waitKey()
```

## INVERTING THE BLUR IMAGE AND CONVERT IMAGE TO PENCIL SKETCH

```python
#invert the blurred image
inverted_blurred=255-blurred
Sketch_pencil=cv2.divide(new_gImage,inverted_blurred,scale=256.0)
cv2.imshow("Sketch",Sketch_pencil)
cv2.waitKey(0)
```

## LOOKING AT BOTH THE ORIGINAL IMAGE AND PENCIL SKETCH 

```python
#look at both original and pencil sketch
cv2.imshow("original image",God_Image)
cv2.imshow("Pencil sketch",Sketch_pencil)
cv2.waitKey(0)
```
